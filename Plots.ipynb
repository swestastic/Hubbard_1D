# Plots for 1D Hubbard Model DQMC

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Mui = 0.0
Muf = 10.0
Mustep = 0.1
Beta = 2.0
U = 4.0
N = 50

Mus = np.arange(Mui, Muf + Mustep, Mustep).round(1)
```

```python
DQMC_Data = np.loadtxt(f"Results/DQMC_Mui{Mus[0]}Muf{Mus[-1]}Mus{Mustep}B{Beta}U{U}N{N}.txt")
n_up_vs_T, n_up_err, n_down_vs_T, n_down_err, double_occ_vs_T, double_occ_err = DQMC_Data
```

```python
SmoQy_Data = np.loadtxt(f"Results/SmoQy_Mui{Mus[0]}Muf{Mus[-1]}Mus{Mustep}B{Beta}U{U}N{N}.txt")
n_up_vs_T_smoqy, n_up_err_smoqy, n_down_vs_T_smoqy, n_down_err_smoqy, double_occ_vs_T_smoqy, double_occ_err_smoqy = SmoQy_Data
```

```python
plt.errorbar(Mus,n_up_vs_T_smoqy+n_down_vs_T_smoqy, yerr=np.sqrt(n_up_err_smoqy**2+n_down_err_smoqy**2), label='SmoQy', color='blue')
plt.errorbar(Mus,n_up_vs_T+n_down_vs_T, yerr=np.sqrt(n_up_err**2+n_down_err**2), label='Python', color='green')
plt.legend()
plt.title(f"Density vs. Chemical Potential, L={N}, Beta={Beta}, U={U}")
plt.xlabel("Chemical Potential (Mu)")
plt.ylabel("Density")
plt.show()

plt.errorbar(Mus,n_up_vs_T_smoqy, yerr=n_up_err_smoqy, label='SmoQy', color='blue')
plt.errorbar(Mus,n_up_vs_T, yerr=n_up_err, label='Python', color='green')
plt.legend()
plt.title(f"Nup vs. Chemical Potential, L={N}, Beta={Beta}, U={U}")
plt.xlabel("Chemical Potential (Mu)")
plt.ylabel("Density")
plt.show()

plt.errorbar(Mus,n_down_vs_T_smoqy, yerr=n_down_err_smoqy, label='SmoQy', color='blue')
plt.errorbar(Mus,n_down_vs_T, yerr=n_down_err, label='Python', color='green')
plt.legend()
plt.title(f"Ndown vs. Chemical Potential, L={N}, Beta={Beta}, U={U}")
plt.xlabel("Chemical Potential (Mu)")
plt.ylabel("Density")
plt.show()

plt.errorbar(Mus,double_occ_vs_T_smoqy, yerr=double_occ_err_smoqy, label='SmoQy', color='blue')
plt.errorbar(Mus,double_occ_vs_T, yerr=double_occ_err, label='Python', color='green')
plt.legend()
plt.title(f"Double Occupancy vs. Chemical Potential, L={N}, Beta={Beta}, U={U}")
plt.xlabel("Chemical Potential (Mu)")
plt.ylabel("Double Occupancy")
plt.show()
```
